```python
#importing libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#set up sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
# add legends
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right')
fig.show()
```

```python
#further modifying our legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right', frameon = True, fontsize=12, ncol = 2, shadow = True)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', bbox_to_anchor=(1,1)) # attach the upper left of the legend to the top right corner of the plot.
#(1,1) = the top right of the plotting area
#(0,0) = the bottom left of the plotting area
```

```python
#texts and  annotations
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

#add texts to specific positions
ax.text(10, 95, "This value is important!")
```

```python

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right', prop={"family":"serif", "size":12})

ax.text(10, 95, "This value is important!", 
        ha='center', 
        color='red', size=20)
```

```python
#positioning text annotations with reference to data points or axes.
fig, ax=plt.subplots() #create emplty plot
ax.axis([0,10,0,10])
ax.text(1,5,".Data:(1,5)", transform=ax.transData)
ax.text(0.5, 0.1, ".Axes:(0.5,0.1)", transform=ax.transAxes)
ax.text(0.5,0.1,".Figure(0.5,0.1)", transform=fig.transFigure)
```

```python
#ANNOTATE WITH ARROWS
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(30, 94), arrowprops=dict(facecolor='black'))
```

```python
ax = plt.sfig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), 
            arrowprops=dict(arrowstyle='<->'), color='hotpink')
```

```python
#removing tick marks and labels
ax = plt.sfig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

#ax.yaxis.set_major_locator(plt.NullLocator())
#ax.xaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())

```

```python
#import and load packages
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#load dataset
tips = sns.load_dataset("tips")
print(tips)
```

```python
#make a plot
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.lineplot(data=tips,
             x='total_bill',
             y='tip')
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(9, 4))
tipgraph=sns.lineplot(data=tips,
             x='total_bill',y='tip',
             color='hotpink',
             linestyle='--',
             linewidth=3,
             marker='o',
            markerfacecolor='indigo' )

tipgraph.set(title='Tips vs. Toatal Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(9, 4))
tipgraph=sns.scatterplot(data=tips,x='total_bill',
                         y='tip',style='time', hue='day',
                         palette = ['purple',
                                    'hotpink', 'deepskyblue',
                                    'yellowgreen'])

tipgraph.set(title='Tips vs. Toatal Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
#pairplots
sns.pairplot(data=tips, hue='day') #good for exploratory analysis
```

```python
#explore specific levels within variables
sns.relplot(data=tips,x="total_bill",y="tip", hue="sex",
            col='day',kind='scatter', col_wrap=2)
```

```python
#qualitative visualizations
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
on_bad_lines='skip')
df
```

```python
text = " ".join(each for each in df.quote)
# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)

# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # remember 'imshow' from when we added pictures to our matplotlib axes
 interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
```

```python
#making venn diagrams
#import library
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
#assign sets
diagram = venn2_unweighted([A, B],
 set_labels = ('Fruits', 'Pies'),
 set_colors=("blue", "red"),
 alpha=0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
```
